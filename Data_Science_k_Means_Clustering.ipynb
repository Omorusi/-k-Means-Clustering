Elbow Techique

```python
k_rng = range(1,10)
sse = []
for k in k_rng:
  km = KMeans(n_clusters=k)
  km.fit(data_cleaned[['Age','Salary']])
  sse.append(km.inertia_)
```

```python
sse
```

```python
plt.xlabel('k')
plt.ylabel('sum of squared error')
plt.plot(k_rng,sse)
```
